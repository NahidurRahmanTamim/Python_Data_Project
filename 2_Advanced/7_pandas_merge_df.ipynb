```python
#libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading data
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
#cleaning
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country']=='United States'].copy()
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short',  aggfunc='size')
df_US_pivot.reset_index(inplace = True)
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
df_US_pivot = df_US_pivot.sort_values('month_no')
df_US_pivot = df_US_pivot.drop(columns='month_no')
df_US_pivot = df_US_pivot.set_index('job_posted_month')
df_US_pivot

```

```python
fd = pd.read_csv("https://lukeb.co/software_csv", index_col=0)
fd
```

```python
df_us_merged=df_US_pivot.merge(fd, on='job_posted_month', how='left')
df_us_merged
```

```python
top_5 = (
    df_us_merged
    .sum()
    .sort_values(ascending=False)
    .head()
    .index.to_list()
    )
top_5
```

```python
df_us_merged[top_5].plot(kind='line')
plt.title('Top 5 tech job demand in the US')
plt.xlabel('2023')
plt.ylabel('Number of job postings')
plt.legend(title='Job Title', bbox_to_anchor=(1, 1.02), loc='upper left')
plt.show()

```
