```python
#libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading data
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
#cleaning
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df_salary = df[pd.notna(df['salary_year_avg'])].copy()
def projected_salary(x):
    return x * 1.03
df_salary['projected salary'] = df_salary['salary_year_avg'].apply(projected_salary)
df_salary[['salary_year_avg', 'projected salary']].head()
```

```python
import ast
def clean_skills(x):
  if pd.notna(x):
    return ast.literal_eval(x)
  else:
    return x

df['job_skills']=df['job_skills'].apply(clean_skills)
```

```python
def projected_salary(Row): # inside the function we can use any name for the row/x/y/etc
    if "Senior" in Row['job_title_short']:
        return Row['salary_year_avg'] * 1.1
    else:
        return Row['salary_year_avg'] * 1.05

df_salary['adjusted salary'] = df_salary.apply(projected_salary, axis=1)
df_salary[['job_title', 'salary_year_avg', 'adjusted salary']].head()
```
