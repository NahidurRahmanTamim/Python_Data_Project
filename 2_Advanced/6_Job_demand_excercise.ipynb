```python
#libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading data
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
#cleaning
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

## Job Demand

```python
df_US = df[df['job_country']=='United States'].copy()
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short',  aggfunc='size')
```

```python
df_US_pivot.reset_index(inplace = True)
```

```python
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
```

```python
df_US = df[df['job_country']=='United States'].copy()
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short',  aggfunc='size')
df_US_pivot.reset_index(inplace = True)
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
df_US_pivot = df_US_pivot.sort_values('month_no')
df_US_pivot = df_US_pivot.drop(columns='month_no')
df_US_pivot = df_US_pivot.set_index('job_posted_month')
df_US_pivot
```

```python
top_3_jobs = df_US['job_title_short'].value_counts().nlargest(3).index.to_list()
```

```python
df_US_pivot[top_3_jobs].plot(kind='line',marker='o')
plt.title('Job Demand Over Months in the US for Top 3 Job Titles')
plt.xlabel('Month')
plt.ylabel('Job Count')

plt.legend(title='Job Title')
plt.show()
```
